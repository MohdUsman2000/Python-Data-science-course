```python
filters=[
    ImageFilter.BLUR,
     ImageFilter.CONTOUR,
      ImageFilter.DETAIL,
       ImageFilter.EDGE_ENHANCE,
        ImageFilter.EDGE_ENHANCE_MORE,
         ImageFilter.EMBOSS,
          ImageFilter.FIND_EDGES,
           ImageFilter.SMOOTH,
            ImageFilter.SMOOTH_MORE,
             ImageFilter.SHARPEN,
]
for i,f in enumerate(filters):
    im2.filter(f).save(f'generated/filter_{i}.png')
```
