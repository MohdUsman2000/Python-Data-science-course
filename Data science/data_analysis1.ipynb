- selecting rows and coloumns
     - rows(slicing)
           - `loc` - gets us the data using named indexes
           - `iloc` -  get us the data using numeric indexes
     - coloumns(dictionary style)
           - `df[single]` - requires name of columns
           - `df[multiple]` -requires a list of coloumns (multiple is list of named cols)

```python
weather.iloc[0] #single row (record)
```

```python
weather.iloc[-1]
```

```python
weather.iloc[25:50]
```

```python
# record 1- 15 with first 5 col
weather.iloc[1:16, :5] #dataframe 
```

```python
#  15 record with last three cols
weather.iloc [:15, -3:] # weather.iloc [-3,-2,-1]
```

```python
weather['Date'] # selcting a single coloumn (dictionary style)-> series
```

```python
cols =[ 'Date', 'Month' , 'Week']
weather[cols] # multiple cols (dictionary style )->> dataframe
```
